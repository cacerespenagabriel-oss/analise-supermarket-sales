### Satisfação e Avaliações

1. Qual é a média geral de avaliação (`Rating`)?
2. Existe correlação entre `Rating` e `Total` (clientes que gastam mais avaliam melhor)?
3. Qual linha de produto tem a maior média de avaliação?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r'C:\Users\USER\OneDrive\Documentos\Repos\Projeto\supermarket_sales.xlsx')

df1 = df.copy()

df1.head()
```

1. Qual é a média geral de avaliação (`Rating`)?

```python
resultado = df1.loc[:, 'Rating'].mean()
display(resultado)
plt.hist(df['Rating'], bins=10, color='plum', edgecolor='black')
plt.title('Distribuição das Avaliações')
plt.xlabel('Avaliação')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

2. Existe correlação entre `Rating` e `Total` (clientes que gastam mais avaliam melhor)?

```python
plt.scatter(df['Total'], df['Rating'], alpha=0.6, color='teal')
plt.title('Correlação entre Total Gasto e Avaliação')
plt.xlabel('Total Gasto')
plt.ylabel('Avaliação')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

3. Qual linha de produto tem a maior média de avaliação?

```python
resultado = df1.loc[:, ['Product line', 'Rating']].groupby('Product line').mean().sort_values('Rating', ascending=False).reset_index()
display(resultado)
```
