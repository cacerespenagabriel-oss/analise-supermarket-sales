### E. Comportamento Temporal

1. Há algum padrão de horário com mais vendas (por exemplo, manhã x tarde)?
2. Quais são as horas do dia mais movimentadas?
3. Há diferença no total de vendas entre dias úteis e fins de semana?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r'C:\Users\USER\OneDrive\Documentos\Repos\Projeto\supermarket_sales.xlsx')
df1 = df.copy()

# Limpando a coluna Date
df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')
df1['Dia_Semana'] = df1['Date'].dt.day_name()
df1['Mes'] = df1['Date'].dt.month_name()
df1['Date'] = df1['Date'].dt.date

display(df1.head())
```

1. Há algum padrão de horário com mais vendas (por exemplo, manhã x tarde)?

```python
resultado = df1.loc[:, ['Time', 'Invoice ID']].groupby('Time').count().sort_values('Invoice ID', ascending=False).reset_index()
display(resultado)
```

3. Há diferença no total de vendas entre dias úteis e fins de semana?

```python
resultado = df1.loc[:, ['Dia_Semana', 'Invoice ID']].groupby('Dia_Semana').count().sort_values('Invoice ID', ascending=False).reset_index()
display(resultado)
```
